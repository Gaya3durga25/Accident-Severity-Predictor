```python
min=list(range(5,56,5))
def convert_minutes(x: int):
    for m in min:
        if x%m==x and x>m-5:
            return m
        if x in [56,57,58,59]:
            return 0
        if x in min+[0]:
            return x
```

```python
df['minute']=df['minute'].apply(lambda x:convert_minutes(x))
```

```python
#null values are categorized into
# 1.MAR(missing at random)
# 2.MNAR(missing not at random)
# 3.MCAR
```

Missing Completely at Random (MCAR)

The missing values have no pattern and don’t depend on any other variable.

Example: A sensor fails randomly, causing missing data.

Missing at Random (MAR)

The missing values depend on some other variables but not on the missing ones themselves.

Example: If income data is missing more for younger people, but age is available, then the missingness is related to age.

Missing Not at Random (MNAR)

The missing values depend on the missing variable itself.

Example: People with high incomes don’t report their salaries because they want privacy.

```python
def target_encoder(df,fi):
    for f in fi:
        f_val=list(np.arange(df[f].nunique()))
        f_key=list(df[f].sort_values().unique())
        f_dict=dict(zip(f_key,f_val))
        df[f]=df[f].map(f_dict)
    return df
df=target_encoder(df,df.drop(['Accident_severity'],axis=1).columns)
df.shape
```

```python
#upsampling using smote
from collections import Counter
from imblearn.over_sampling import SMOTE
counter=Counter(y_train)
print("================================================")
for k,v in counter.items():
    per=100*v/len(y_train)
    print(f"class= {k}, n={v} ({per:.2f}%)")
oversample=SMOTE()
x_train,y_train=oversample.fit_resample(x_train,y_train)

counter=Counter(y_train)

print("=================================================")

for k,v in counter.items():
    per=100*v/len(y_train)
    print(f"class= {k}, n={v} ({per:.2f}%)")
print("===========================================")
print("Upsampled data shape :",x_train.shape,y_train.shape)
```

```python
def modelling(X_train, y_train, X_test, y_test, **kwargs):
    scores = {}
    models = []
    bvd = {}
    if 'xgb' in kwargs.keys() and kwargs['xgb']:
        xgb = XGBClassifier()
        xgb.fit(X_train._get_numeric_data(), np.ravel(y_train, order='C'))
        y_pred = xgb.predict(X_test._get_numeric_data())
        scores['xgb']= [accuracy_score(y_test, y_pred)]
        models.append(xgb)


    if 'rf' in kwargs.keys() and kwargs['rf']:
        rf = RandomForestClassifier(n_estimators=200)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_test)
        scores['rf']= [accuracy_score(y_test, y_pred)]
        models.append(rf)

    if 'extree' in kwargs.keys() and kwargs['extree']:
        extree = ExtraTreesClassifier()
        extree.fit(X_train, y_train)
        y_pred = extree.predict(X_test)
        scores['extree'] = [accuracy_score(y_test, y_pred)]
        models.append(extree)
        
    return scores,models
```

```python

from sklearn.model_selection import cross_val_score  # ✅ Correct

from sklearn.model_selection import KFold  
from sklearn.model_selection import GridSearchCV

gkf = KFold(n_splits=3, shuffle=True, random_state=42).split(X=x_train, y=y_train)

# A parameter grid for ETrees
params = {
    'n_estimators': range(100, 500, 100),  # [100,200,300,400,500]
    'ccp_alpha': [0.0, 0.1],
    'criterion': ['gini'],
    'max_depth': [5,11],
    'min_samples_split': [2,3],
}

extree_estimator = ExtraTreesClassifier()

gsearch = GridSearchCV(
    estimator= extree_estimator,
    param_grid= params,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=gkf,
    verbose=1,
)


extree_model = gsearch.fit(X=x_train, y=y_train)
(gsearch.best_params_, gsearch.best_score_)
```
